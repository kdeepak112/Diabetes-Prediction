```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('diabetes_data_upload.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
```

```python
columns = df.columns[1:]
```

```python
columns = columns.tolist()
```

```python
columns
```

```python
for i in columns:
    df[i] = le.fit_transform(df[i])
```

```python
len(df.loc[df['class'] == 1])
```

```python
df.corr()
```

```python
import seaborn as sns
```

```python
sns.jointplot(x='Age',y='class',data=df)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
rf = RandomForestClassifier()
knn = KNeighborsClassifier(n_neighbors=5)
svm = svm.SVC()
nb = GaussianNB()
```

```python
classifiers = {rf:0,knn:0,svm:0,nb:0}
```

```python
x = df[df.columns[:-1]]
```

```python
x
```

```python
y = df['class']
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in classifiers:
    score = cross_val_score(i,x,y,cv=5).mean()
    classifiers[i] = score
```

```python
classifiers
```

```python
n_estimators = [int(x) for x in np.linspace(10,100,10)]
max_features = ['auto','sqrt']
max_depth = [2,4,6,8,10]
min_samples_split = [2,5,7,9,11]
min_samples_leaf = [1,2,3,4]
bootstrap = [True,False]
```

```python
param_grid = {
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
    'bootstrap':bootstrap
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_grid = GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,verbose=1,n_jobs=4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Up Sampling
```

```python
from imblearn.combine import SMOTETomek
smk = SMOTETomek(random_state = 42)
x_res,y_res = smk.fit_sample(x,y)
```

```python
x_res = pd.DataFrame(x_res,columns=x.columns)
y_res = pd.DataFrame(y_res,columns=['class'])
df_up = pd.concat([x_res,y_res],axis=1)
```

```python
df_up
```

```python
print(len(df_up.loc[df_up['class'] == 0]))
print(len(df_up.loc[df_up['class'] == 1]))
```

```python
x_up = df_up[df_up.columns[:-1]]
```

```python
y_up = df_up['class']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_up,y_up,test_size=0.25,random_state=101)
```

```python
from sklearn.feature_selection import mutual_info_classif
# determine the mutual information
mutual_info = mutual_info_classif(xtrain, ytrain)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = xtrain.columns
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import SelectKBest
sel_five_cols = SelectKBest(mutual_info_classif, k=5)
sel_five_cols.fit(xtrain, ytrain)
column = xtrain.columns[sel_five_cols.get_support()]
column = column.tolist()
column
```

```python
X_new = df_up[column]
X_new.head()
```

```python
Y_new = df_up['class']
Y_new.head()
```

```python
Xnew_train, Xnew_test, ynew_train, ynew_test = train_test_split(X_new,Y_new,test_size=0.25,random_state=101)
```

```python
Xnew_test
```

```python
ynew_test
```

```python
rf_grid.fit(Xnew_train,ynew_train)
```

```python
rf_grid.best_params_
```

```python
rf = RandomForestClassifier(bootstrap= False,
 max_depth= 10,
 max_features= 'auto',
 min_samples_leaf= 1,
 min_samples_split= 5,
 n_estimators= 10)
```

```python
rf.fit(Xnew_train,ynew_train)
```

```python
pred_n = rf.predict(Xnew_test)
```

```python
rf.score(Xnew_test,ynew_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(ynew_test,pred_n))
print(confusion_matrix(ynew_test,pred_n))
```
